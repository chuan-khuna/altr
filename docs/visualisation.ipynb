```python
from altr.visualisation import set_style, set_seaborn_palette
from altr import visualisation as altr_vis

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import patheffects
```

```python
altr_vis.available_palettes
```

```python
altr_vis.available_styles
```

generate mock data


```python
n = 100
num_series = 5

df = pd.DataFrame(
    np.random.randn(n, num_series),
    index=np.arange(n),
    columns=[f'fake {i + 1}' for i in range(num_series)],
)

df = df.cumsum()

df
```

```python
figsize = (8, 4)

for style in altr_vis.available_styles:
    set_style(style)
    set_seaborn_palette(style)

    fig = plt.figure(figsize=figsize, dpi=100)

    for col in df.columns:
        sns.lineplot(
            x=df.index,
            y=df[col],
            path_effects=[patheffects.withStroke(linewidth=3, foreground="#F2DFCE")],
        )

    plt.title(style)
    plt.xlabel('')
    plt.ylabel('')
    plt.show()
```

```python
set_style('ft')
set_seaborn_palette('ft', 'vis_colors')

fig = plt.figure(figsize=figsize, dpi=100)

for col in df.columns:
    sns.lineplot(
        x=df.index,
        y=df[col],
        path_effects=[patheffects.withStroke(linewidth=3, foreground="#F2DFCE")],
    )

plt.title('ft')
plt.xlabel('')
plt.ylabel('')
plt.show()
```

```python
set_style('ft')
set_seaborn_palette('ft', 'main_colors')

fig = plt.figure(figsize=figsize, dpi=100)

for col in df.columns:
    sns.lineplot(
        x=df.index,
        y=df[col],
        path_effects=[patheffects.withStroke(linewidth=3, foreground="#F2DFCE")],
    )

plt.title('ft')
plt.xlabel('')
plt.ylabel('')
plt.show()
```
